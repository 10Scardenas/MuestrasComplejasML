```python
from sklearn.preprocessing import StandardScaler

#Definimos las moleculas como Features
dt_features= dt_spec_cut.to_numpy() #Variables predictoras
dt_features= StandardScaler().fit_transform(dt_features)
#Dfenimos CCR% como Target
dt_target=dt_Supp_Prop['CCR (wt%)'].to_numpy() #Variable Objetivo

dt_features = StandardScaler().fit_transform(dt_features)


#partir conjunto de entrenamiento
X_train, X_test, y_train, y_test= train_test_split(dt_features, dt_target, test_size=0.3,random_state=42) #parametro random state permite que siempre su usen las mismas muestras

print(X_train.shape)
print(y_train.shape)
```

```python
pca = PCA()
pca.fit(X_train)

plt.plot(range(len(pca.explained_variance_)), pca.explained_variance_ratio_)
plt.show()
```
