```python
# Predicciones test
# ==============================================================================
predicciones1 = modelo1.predict(X_te1)
predicciones1
```

## **Metricas**

```python
# Accuracy de test del modelo 
# ==============================================================================
accuracy = accuracy_score(
            y_true    = y_te1,
            y_pred    = predicciones1,
            normalize = True
           )
print("")
print(f"El accuracy de test es: {100*accuracy}%")
```
